```python
X= df.drop('Class',axis=1)
```

```python
y=df['Class']
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, precision_recall_curve, accuracy_score, ConfusionMatrixDisplay
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
## Define the function for the model
```

```python
def models_prediction(model, X_train, X_test, y_train, y_test):
    # Train the model
    model.fit(X_train, y_train)
    
    # Make predictions
    pred = model.predict(X_test)
    
    # Evaluate performance
    print('Accuracy:', accuracy_score(y_test, pred))
    print(classification_report(y_test, pred))
    
    # Compute confusion matrix
    matrix = confusion_matrix(y_test, pred)
    
    # Plot confusion matrix
    disp = ConfusionMatrixDisplay(confusion_matrix=matrix, display_labels=model.classes_)
    disp.plot(cmap=plt.cm.Blues)
    plt.show()
```

```python
model_Random= RandomForestClassifier()
```

```python
models_prediction( model_Random,  X_train, X_test, y_train, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
Decision_tree= DecisionTreeClassifier()
```

```python
models_prediction (Decision_tree, X_train, X_test, y_train, y_test)
```
